```python
import requests as r
import json
import pandas as pd
from many_requests import ManyRequests
import numpy as np
import csv
import ast
```

```python
df_movies = pd.read_csv("./created/movies.csv")
```

```python
df_movies
```

```python
correct_format_studio = df_movies["studios"].apply(lambda x: ast.literal_eval(x))
correct_format_studio = correct_format_studio.apply(lambda x: ":".join(x))
correct_format_genres = df_movies["genres"].apply(lambda x: ast.literal_eval(x))
correct_format_genres = correct_format_genres.apply(lambda x: ":".join(x))
```

```python
df_movies["studios"] = correct_format_studio
df_movies["genres"] = correct_format_genres
```

```python
df_movies.to_csv("./created/correct_format_movies.csv", index = False)
```
